````
from numpy.random import multivariate_normal
c = np.array([[3, 2], [2, 3]])
c
````

````
multivariate_normal(mean=[50, 10],cov=c, size=4)
````

````
vals = multivariate_normal([50, 10], c, 200)
data_x = [x for (x, y) in vals]
data_y = [y for (x, y) in vals]
plt.scatter(data_x, data_y)
````
